A point inside a pentagon repeatedly jumps half of the distance towards a randomly chosen vertex, but the currently chosen vertex cannot be the same as the previously chosen vertex.

<img src="Pentagon_factral_1.png" width="500"/>

```python
from turtle import *
import random
import math
import turtle

p = []
radius = 300
for i in range(5):
    angle = math.radians(72 * i - 90)  
    x = radius * math.cos(angle)
    y = radius * math.sin(angle)
    p.append([x, y])

tracer(100)
hideturtle()
delay(1)

index = 0
for x, y in p:
    penup()
    goto(x, y)
    pendown()
    dot(2)

x = 0
y = 0
poly = 5  

for i in range(100000):
    prev_index = index
    index = random.randint(0, 4)
    
    
    if (poly + index - prev_index) % poly != 1:
        vertex = p[index]
        x = (x + vertex[0]) / 2
        y = (y + vertex[1]) / 2
        
        if i % 2000 == 0:
            print(i, "iterations done")
        
        penup()
        goto(x, y)
        pendown()
        dot(2)
        penup()

done()
```

A point inside a pentagon repeatedly jumps half of the distance towards a randomly chosen vertex, but the currently chosen vertex cannot neighbor the previously chosen vertex if the two previously chosen vertices are the same.

<img src="Pentagon_factral_5.png"  width="500"/>

```python
from turtle import *
import random
import math

p = []
radius = 300
for i in range(5):
    angle = math.radians(72 * i - 90)  
    x = radius * math.cos(angle)
    y = radius * math.sin(angle)
    p.append([x, y])

tracer(100)
hideturtle()
delay(1)

# Draw the vertices
for x, y in p:
    penup()
    goto(x, y)
    pendown()
    dot(2)

x = 0
y = 0
prev_index = 0
second_prev_index = 0
poly = 5  

for i in range(100000):
    index = random.randint(0, poly-1)
    
    # If last two vertices were the same and new vertex is adjacent, choose again
    while (prev_index == second_prev_index and 
           (poly + index - prev_index) % poly == 1 or 
           (poly + prev_index - index) % poly == 1):
        index = random.randint(0, poly-1)
    
    second_prev_index = prev_index
    prev_index = index
    
    vertex = p[index]
    x = (x + vertex[0]) / 2
    y = (y + vertex[1]) / 2
    
    if i % 2000 == 0:
        print(i, "iterations done")
    
    penup()
    goto(x, y)
    pendown()
    dot(2)
    penup()

done()
```

```python
from turtle import *
import random
import math

p = []
radius = 300
for i in range(5):
    angle = math.radians(72 * i - 90) 
    x = radius * math.cos(angle)
    y = radius * math.sin(angle)
    p.append([x, y])

p.append([0, 0])

tracer(100)
hideturtle()
delay(1)

for x, y in p:
    penup()
    goto(x, y)
    pendown()
    dot(2)

x = 0
y = 0
prev_index = 0
second_prev_index = 0
poly = 5  

for i in range(100000):
    index = random.randint(0, poly)  
    
    # If last two vertices were the same and new vertex is adjacent (only check for pentagon vertices)
    while (prev_index < poly and second_prev_index < poly and  # Only apply rule for pentagon vertices
           prev_index == second_prev_index and 
           index < poly and  # Only check adjacency for pentagon vertices
           ((poly + index - prev_index) % poly == 1 or 
            (poly + prev_index - index) % poly == 1)):
        index = random.randint(0, poly)
    
    second_prev_index = prev_index
    prev_index = index
    
    vertex = p[index]
    x = x + (vertex[0] - x) * 2/3  
    y = y + (vertex[1] - y) * 2/3  
    
    if i % 2000 == 0:
        print(i, "iterations done")
    
    penup()
    goto(x, y)
    pendown()
    dot(2)
    penup()

done()
```
