```python
from turtle import *
import random
import math

p = []
radius = 300

# Create 8 vertices for an octagon (45 degrees apart)
for i in range(8):
    angle = math.radians(45 * i - 90)
    x = radius * math.cos(angle)
    y = radius * math.sin(angle)
    p.append([x, y])

# Add center point
p.append([0, 0])

# Create a list of colors for each vertex (8 colors for vertices + 1 for center)
colors = [
    (255, 0, 0),      # Red (0)
    (255, 127, 0),    # Orange (1)
    (255, 255, 0),    # Yellow (2)
    (0, 255, 0),      # Green (3)
    (0, 255, 255),    # Cyan (4)
    (0, 0, 255),      # Blue (5)
    (127, 0, 255),    # Purple (6)
    (255, 0, 255),    # Magenta (7)
    (200, 200, 200),  # Light gray (center - 8)
]

tracer(100)
hideturtle()
delay(1)
colormode(255)  # Set colormode to use RGB values

# Draw all vertices with their respective colors
for i, (x, y) in enumerate(p):
    penup()
    goto(x, y)
    pendown()
    pencolor(colors[i])
    dot(5)  # Slightly larger dot for vertices

x = 0
y = 0
prev_index = 0
second_prev_index = 0
poly = 8  # Number of outer vertices
total_vertices = poly + 1  # Including center

for i in range(100000):
    # Random index including center (0 to 8)
    index = random.randint(0, total_vertices - 1)
    
    # Apply adjacent vertex restriction only for outer vertices
    if prev_index < poly and second_prev_index < poly and prev_index == second_prev_index and index < poly:
        while ((poly + index - prev_index) % poly == 1 or (poly + prev_index - index) % poly == 1):
            index = random.randint(0, total_vertices - 1)  # Try a new random vertex (including center)
    
    # Update history
    second_prev_index = prev_index
    prev_index = index
    
    # Get the chosen vertex coordinates
    vertex = p[index]
    
    # Move 2/3 of the way toward the chosen vertex
    x = x + (vertex[0] - x) * 2/3 
    y = y + (vertex[1] - y) * 2/3
    
    # Set color based on the chosen vertex (where we're jumping TO)
    pencolor(colors[index])
    
    if i % 2000 == 0:
        print(i, "iterations done")
    
    # Draw the dot at the new position
    penup()
    goto(x, y)
    pendown()
    dot(2)
    penup()

done()
```

```python
from turtle import *
import random
import math
import colorsys

# Set up the screen
setup(800, 800)
bgcolor("black")
tracer(100)
hideturtle()
delay(1)
colormode(255)

# Create vertices of a 100-sided polygon
p = []
radius = 300
num_vertices = 100

# Create outer vertices
for i in range(num_vertices):
    angle = math.radians(360/num_vertices * i)
    x = radius * math.cos(angle)
    y = radius * math.sin(angle)
    p.append([x, y])

# Generate colors for each vertex
colors = []
for i in range(num_vertices):
    # Use HSV color space to generate evenly spaced colors around the color wheel
    h = i / num_vertices
    s = 1.0
    v = 1.0
    # Convert HSV to RGB
    r, g, b = colorsys.hsv_to_rgb(h, s, v)
    # Scale to 0-255
    colors.append((int(r*255), int(g*255), int(b*255)))

# Draw all vertices with their respective colors
for i, (x, y) in enumerate(p):
    penup()
    goto(x, y)
    pendown()
    pencolor(colors[i])
    dot(2)

# Starting position
x = 0
y = 0
prev_index = 0
second_prev_index = 0
poly = num_vertices  # Number of outer vertices

# Main chaos game loop
for i in range(100000):
    # Random index for outer vertices only (0 to num_vertices-1)
    index = random.randint(0, poly - 1)
    
    # Apply adjacent vertex restriction 
    if prev_index == second_prev_index and ((poly + index - prev_index) % poly == 1 or (poly + prev_index - index) % poly == 1):
        while ((poly + index - prev_index) % poly == 1 or (poly + prev_index - index) % poly == 1):
            index = random.randint(0, poly - 1)
    
    # Update history
    second_prev_index = prev_index
    prev_index = index
    
    # Get the chosen vertex coordinates
    vertex = p[index]
    
    # Move 2/3 of the way toward the chosen vertex
    x = x + (vertex[0] - x) * 3/5
    y = y + (vertex[1] - y) * 3/5
    
    # Set color based on the chosen vertex
    pencolor(colors[index])
    
    if i % 2000 == 0:
        print(i, "iterations done")
    
    # Draw the dot at the new position
    penup()
    goto(x, y)
    pendown()
    dot(1)  # Smaller dot size for cleaner appearance with many points
    penup()

done()
```
