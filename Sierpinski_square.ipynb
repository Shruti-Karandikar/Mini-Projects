If the chaos game is run with a square, no fractal appears and the interior of the square fills evenly with points. 

However, if restrictions are placed on the choice of vertices, fractals will appear in the square. Here are some examples:

<img src="Fartorial_1.png" alt="dmin and max" width="500"/>

```python
# If the current vertex cannot be chosen in the next iteration, this fractal appears:
p = [[-300,-300], [300,-300], [300, 300], [-300, 300]]
tracer(100)
hideturtle()
delay(1)
index = 0

for x, y in p:
    penup()
    goto(x, y)
    pendown()
    dot(2)

for i in range(100000):
    prev_index = index
    index = random.randint(0, 3)
    if index != prev_index:
        vertex = p[index]
        x = (x + vertex[0]) / 2
        y = (y + vertex[1]) / 2
    
        if i % 2000 == 0:
            print(i, "iterations done")
    
        penup()
        goto(x, y)
        pendown()
        dot(2)
        penup()
    
done()
```

If the current vertex cannot be one place away (anti-clockwise) from the previously chosen vertex, this fractal appears:

<img src="Factorial_2.png" alt="dmin and max" width="500"/>


```python
p = [[-300,-300], [300,-300], [300, 300], [-300, 300]]
tracer(100)
hideturtle()
delay(1)

index = 0
for x, y in p:
    penup()
    goto(x, y)
    pendown()
    dot(2)

poly = 4
for i in range(100000):
    prev_index = index
    index = random.randint(0, 3)
    
    if (poly + index - prev_index)%poly != 1:  
        vertex = p[index]
        x = (x + vertex[0]) / 2
        y = (y + vertex[1]) / 2
        
        if i % 2000 == 0:
            print(i, "iterations done")
        
        penup()
        goto(x, y)
        pendown()
        dot(2)
        penup()
            
done()
```

When the jump is 2/3 and the point can also jump towards the center of the square, the chaos game generates the Vicsek fractal:

 <img src="Factral_5.png" alt="dmin and max" width="500"/>

```python
import turtle
import random

p = [[0,0], [-300,-300], [300,-300], [300, 300], [-300, 300]]
tracer(100)
hideturtle()
delay(1)

x, y = p[0]
current_index = 0
prev_index = 0

for vx, vy in p:
    penup()
    goto(vx, vy)
    pendown()
    dot(2)

for i in range(100000):
    last_index = current_index
    current_index = random.randint(0, 4)
    
    if last_index != prev_index or (current_index - last_index) % 5 not in [1, 4]:
        vertex = p[current_index]
        x = (x + vertex[0]) / 2
        y = (y + vertex[1]) / 2
        
        if i % 2000 == 0:
            print(i, "iterations done")
        
        penup()
        goto(x, y)
        pendown()
        dot(2)
        
        prev_index = last_index
done()
```

Making my own fractal:

A point inside a square repeatedly jumps half of the distance towards a randomly chosen vertex, but consecutive vertex choices must have a remainder of 0 when their difference is divided by 2. This creates a distinct pattern as the point traces its path, with some vertex combinations forbidden due to this modulo restriction.

 <img src="Factral_6.png" alt="dmin and max" width="500"/>

```python
p = [[-300,-300], [300,-300], [300, 300], [-300, 300]]
tracer(100)
hideturtle()
delay(1)

index = 0
for x, y in p:
    penup()
    goto(x, y)
    pendown()
    dot(2)

poly = 2
for i in range(100000):
    prev_index = index
    index = random.randint(0, 3)
    
    if (poly + index - prev_index)%poly != 1:  
        vertex = p[index]
        x = (x + vertex[0]) / 2
        y = (y + vertex[1]) / 2
        
        if i % 2000 == 0:
            print(i, "iterations done")
        
        penup()
        goto(x, y)
        pendown()
        dot(2)
        penup()
            
done()
```
