collatz conjecture:

Take any positive integer n. If the term is even, divide by two. If the term is odd, multiply it by three and add one. we repeat this process till we get a 1.

## Part 1

```python
# Returns the number of steps
def collatz(n):
    counter = 0
    if n < 1:
        return counter
    while n != 1:
        counter = counter + 1
        if n % 2 == 0:
            n = n//2
        else:
            n = 3 * n + 1   
    return counter
```

I have plotted the number of steps for the first 10, 100 and 1000 numbers to see any patterens that may exist.

```python
count = []
for i in range(1000):
    count.append(collatz(i))
#print(count)
```

```python
# For 10 numbers
plt.plot(count, ".")
```

```python
# For 100 numbers
plt.plot(count, ".")
```

```python
# For 1000 numbers
plt.plot(count, ".")
```

## Part 2

We have looked at the paath lengths now lets look at the individual path.

```python
# Collatz path
def collatz_path(n):
    path = []
    while n != 1:
        if n % 2 == 0:
            n = n//2
        else : 
            n = 3*n + 1
        path.append(n)
    return(path)
```

```python
plt.plot(collatz_path(23))
```

We see that the path is not smooth and it can bounce around a lot. There doesn't seem to be a pattern - some numbers go down to 1 immediately, others take their own sweet time.

## Part 3

In the original plot, there were 1000 points but the path lengths go only upto 175. This is an example of the pigeon hole principle- many numbers have the same path length. 

How can we visualize this? Is there a pattern?

```python
def analyze_path(start, end):
    frequencies = {}
    for i in range(start, end):
        length = collatz(i)
        if length in frequencies:
            frequencies[length] = frequencies[length] + 1
        else:
            frequencies[length] = 1

    return frequencies
```

```python
f = analyze_path(1, 1000)
print(type(f))
```

```python
plt.plot(f.keys(), f.values())
```

As the dictionary keys are not sorted the plot is all over the place.

```python
plt.scatter(f.keys(), f.values())
```

With a scatter plot, its much more understandable. We can relate the data here with the path lenght plot abouve. In particular, observe the three points in the bottom right. They correspond to the long length paths that appear on the top in the original collatz plot.

It seems like path lenghts around 25 and 125 occur more frequntly than the others.
