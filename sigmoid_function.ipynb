A sigmoid function is an S-shaped mathematical function that maps any real number input to an output between 0 and 1. The most common form is:
$Ïƒ(x) = 1 / (1 + e^(-x))$

The sigmoid function is widely used in:

- Machine Learning:
In logistic regression as the activation function to convert linear predictions into probability values (0-1).
In neural networks as an activation function. 

- Classification Problems:
The smooth transition between 0 and 1 makes it ideal for binary classification.
The output can be interpreted as a probability

- Signal Processing:
For smoothing signal transitions

```python
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(x):
    return 1.0 / (1.0 + np.exp(-x))

x = np.linspace(-10, 10, 100)
y = sigmoid(x) * 10

plt.figure(figsize=(5, 5))
plt.plot(x, y)
plt.show()
```

- As x approaches negative infinity, the output approaches 0
- As x approaches positive infinity, the output approaches 1
- The function has a smooth, continuous transition between these values
- The steepest part of the curve is at x=0, where the output is exactly 0.5
- `elev` (elevation): Controls the vertical viewing angle in degrees
- `azim` (azimuth): Controls the horizontal rotation angle in degrees

```python
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(x):
    return 1.0 / (1.0 + np.exp(-x))

x = np.linspace(-10, 10, 100)
y = np.linspace(-10, 10, 100)
X, Y = np.meshgrid(x, y)
Z = sigmoid(X)

fig = plt.figure(figsize=(6, 6))
ax = fig.add_subplot(111, projection='3d')
surf = ax.plot_surface(X, Y, Z, cmap='viridis')

ax.view_init(elev=30, azim=240)

plt.show()
```
