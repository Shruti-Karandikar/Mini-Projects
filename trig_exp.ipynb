```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#sintheta
theta = np.linspace(0, 2 * np.pi, 1000)
sin_theta = np.sin(theta) 
plt.plot(sin_theta)
plt.grid(True)
```

```python
#costheta
cos_theta = np.cos(theta)
plt.plot(cos_theta)
plt.grid(True)
```

```python
plt.plot(2*sin_theta*cos_theta)
plt.grid(True)
```

When plotting sine and cosine waves, each exhibits distinct characteristics: 

sine has zeros at 0, π, and 2π radians, while reaching peaks of ±1 at π/2 and 3π/2 radians.

cosine, shifted π/2 ahead, has zeros at π/2 and 3π/2 radians, with peaks of ±1 at 0, π, and 2π radians. 

These waves intersect at π/4, 3π/4, 5π/4, and 7π/4 radians (within a single cycle), where both equal $1/\sqrt{2}$. When examining their product sin(θ)cos(θ), a fascinating transformation emerges: zeros occur wherever either function is zero (at 0, π/2, π, 3π/2, and 2π radians), while the amplitude reaches only 1/2 at π/4, 3π/4, 5π/4, and 7π/4 radians. 

Most notably, while sin(θ) and cos(θ) each complete one full cycle over 2π radians, their product completes two complete cycles in the same range—effectively doubling the frequency while halving the amplitude. This relationship is captured in the identity sin(θ)cos(θ) = (1/2)sin(2θ), where the coefficient 1/2 represents the amplitude reduction and 2θ indicates the frequency doubling, demonstrating how multiplication in the time domain creates new frequency components.

$2\sin(\theta)\cos(\theta$)
- This is composed of two separate functions, each with the fundamental frequency
- $\sin(\theta)$ has frequency $1$ (completes one cycle as $\theta$ goes from $0$ to $\pi$)
- $\cos(\theta)$ has frequency $1$ as well
- Their multiplication produces the higher frequency


```python
plt.plot(np.sin(2*theta))
plt.grid(True)
```

$\sin(2\theta)$
- This is directly a sine wave with twice the base frequency
- The "$2\theta$" indicates the function completes two full cycles when $\theta$ goes from $\theta$ to $2\pi$
- This representation explicitly shows the doubled frequency

$2\sin(\theta)\cos(\theta)$ demonstrates frequency mixing:

When two signals of frequencies $f_1$ and $f_2$ are multiplied, they produce components at the sum $(f_1 + f_2)$ and difference $(f_1 - f_2)$ of their frequencies.
Here, $\sin(\theta)$ and $\cos(\theta)$ both have frequency $1$, so their product contains:
- A component at frequency $1 + 1 = 2$
- A component at frequency $1 - 1 = 0$
Using the identity $\sin(A)\cos(B)$ = $\frac{1}{2}[\sin(A+B) + \sin(A-B)]$:
$2\sin(\theta)\cos(\theta) = \sin(2\theta) + \sin(\theta) = \sin(2\theta)$

```python
plt.plot(np.cos(theta)**2 - np.sin(theta)**2)
plt.grid(True)
```

$\cos^2(\theta) - sin^2(\theta)$
- Contains squared terms of the fundamental frequency components
- When we square trigonometric functions, we generate new frequency components:

$\cos^2(\theta) = (1 + \cos(2\theta))/2$

$\sin^2(\theta) = (1 - \cos(2\theta))/2$

Subtracting these squared terms:

$\cos^2(\theta) - \sin^2(\theta) = (1 + \cos(2\theta))/2 - (1 - \cos(2\theta))/2$

$= cos(2\theta)$

When you square a sinusoid, you create:
- A constant (DC) component
- A component at twice the original frequency
- In $cos^2(\theta) - sin^2(\theta)$:

The constant components ($1/2$ terms) cancel out

The double-frequency components ($\cos(2\theta)/2$ terms) add together

These forms show how the squared terms must be scaled and shifted to produce the double-frequency cosine wave.
